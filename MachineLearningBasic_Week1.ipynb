# Lí thuyết về Machine Learning

## 1. Giới Thiệu

### 1.1. Machine Learning (ML) là gì? Supervised vs Unsupervised vs Reinforcement Learning
- **Machine Learning (ML)** là một nhánh của trí tuệ nhân tạo (AI), tập trung vào việc phát triển các thuật toán cho phép máy tính học hỏi từ dữ liệu mà không cần lập trình rõ ràng cho từng nhiệm vụ.
- **Supervised Learning (Học có giám sát):** Là phương pháp mà mô hình được huấn luyện trên tập dữ liệu có nhãn (labeled data), bao gồm đầu vào (input) và đầu ra mong muốn (output). Ví dụ: Phân loại (classification) và hồi quy (regression).
- **Unsupervised Learning (Học không giám sát):** Là phương pháp mà mô hình làm việc với dữ liệu không có nhãn, nhằm tìm ra cấu trúc hoặc mẫu ẩn trong dữ liệu. Ví dụ: Phân cụm (clustering) và giảm chiều dữ liệu (dimensionality reduction).
- **Reinforcement Learning (Học tăng cường):** Là phương pháp mà một tác nhân (agent) học cách đưa ra quyết định bằng cách tương tác với môi trường, nhận phần thưởng (reward) hoặc hình phạt (penalty) dựa trên hành động. Ví dụ: Đào tạo AI chơi game hoặc robot tự động điều hướng.
### 1.2. Các thư viện python quan trọng cho Machine Learning
- **NumPy:** Thư viện cơ bản cho tính toán số học, hỗ trợ mảng đa chiều (arrays).
- **Pandas:** Thư viện mạnh mẽ để xử lý và phân tích dữ liệu với cấu trúc DataFrame.
- **Matplotlib:** Thư viện trực quan hóa dữ liệu, vẽ biểu đồ 2D.
- **Seaborn:** Xây dựng trên Matplotlib, cung cấp giao diện trực quan hóa thống kê đẹp mắt.
- **Scikit-Learn:** Thư viện học máy phổ biến, có sẵn một vài tập datasets, hỗ trợ phân loại, hồi quy, gom cụm.

## 1. NumPy
Thư viện **NumPy** được dùng để xử lý mảng đa chiều và tính toán số học, là nền tảng cho các phép toán trong ML.

### 1.1. Tạo và thao tác với mảng
- `np.array(list)`: Tạo mảng NumPy từ list.
- `np.zeros(shape)`: Tạo mảng toàn số 0 với kích thước `shape`.
- `np.ones(shape)`: Tạo mảng toàn số 1 với kích thước `shape`.
- `np.arange(start, stop, step)`: Tạo mảng với dãy số từ `start` đến `stop`, bước nhảy `step`.
- `np.linspace(start, stop, num)`: Tạo mảng với `num` giá trị cách đều nhau từ `start` đến `stop`.
- `np.random.rand(shape)`: Tạo mảng với các giá trị ngẫu nhiên từ 0 đến 1.
- `np.random.randint(low, high, size)`: Tạo mảng với các số nguyên ngẫu nhiên từ `low` đến `high`.

### 1.2. Thao tác trên mảng
- `array.reshape(shape)`: Thay đổi hình dạng của mảng.
- `array.T`: Chuyển vị mảng (transpose).
- `np.concatenate((array1, array2), axis)`: Ghép hai mảng theo trục `axis`.
- `np.vstack((array1, array2))`: Ghép hai mảng theo chiều dọc.
- `np.hstack((array1, array2))`: Ghép hai mảng theo chiều ngang.
- `np.split(array, indices_or_sections)`: Chia mảng thành nhiều phần.

### 1.3. Tính toán số học
- `np.sum(array, axis)`: Tính tổng các phần tử trong mảng theo trục `axis`.
- `np.mean(array, axis)`: Tính trung bình các phần tử.
- `np.std(array, axis)`: Tính độ lệch chuẩn.
- `np.min(array)`, `np.max(array)`: Tìm giá trị nhỏ nhất/lớn nhất.
- `np.dot(array1, array2)`: Tích vô hướng giữa hai mảng.
- `np.linalg.inv(array)`: Tính ma trận nghịch đảo.

## 2. Pandas
Thư viện **Pandas** được dùng để xử lý và phân tích dữ liệu, đặc biệt với cấu trúc **DataFrame**.

### 2.1. Tạo và đọc dữ liệu
- `pd.DataFrame(data)`: Tạo DataFrame từ dữ liệu (list, dict, mảng NumPy, v.v.).
- `pd.read_csv(file_path)`: Đọc file CSV vào DataFrame.
- `pd.read_excel(file_path)`: Đọc file Excel vào DataFrame.
- `df.to_csv(file_path)`: Xuất DataFrame ra file CSV.

### 2.2. Khám phá dữ liệu
- `df.head(n)`: Xem `n` dòng đầu của DataFrame (mặc định là 5).
- `df.tail(n)`: Xem `n` dòng cuối của DataFrame.
- `df.info()`: Hiển thị thông tin tổng quan (cột, kiểu dữ liệu, giá trị null).
- `df.describe()`: Thống kê cơ bản (mean, std, min, max, v.v.).
- `df.shape`: Kích thước của DataFrame (số dòng, số cột).
- `df.columns`: Lấy danh sách tên cột.
- `df.dtypes`: Kiểm tra kiểu dữ liệu của từng cột.

### 2.3. Làm sạch dữ liệu
- `df.drop(columns=[col1, col2])`: Xóa cột khỏi DataFrame.
- `df.dropna()`: Xóa các dòng chứa giá trị `NaN`.
- `df.fillna(value)`: Thay thế `NaN` bằng `value`.
- `df.duplicated()`: Kiểm tra các dòng trùng lặp.
- `df.drop_duplicates()`: Xóa các dòng trùng lặp.
- `df.rename(columns={old: new})`: Đổi tên cột.

### 2.4. Lọc và truy vấn dữ liệu
- `df[col]`: Lấy cột `col` dưới dạng Series.
- `df[[col1, col2]]`: Lấy nhiều cột dưới dạng DataFrame.
- `df.iloc[row_idx, col_idx]`: Truy cập dữ liệu theo chỉ số.
- `df.loc[row_label, col_label]`: Truy cập dữ liệu theo nhãn.
- `df[df[col] > value]`: Lọc dữ liệu dựa trên điều kiện.
- `df.groupby(col)`: Nhóm dữ liệu theo cột `col`.
- `df.merge(df2, on=col)`: Kết hợp hai DataFrame dựa trên cột `col`.

## 5. Scikit-Learn
Thư viện **Scikit-Learn** cung cấp các công cụ để xây dựng, huấn luyện và đánh giá mô hình ML.

### 5.1. Tiền xử lý dữ liệu
- `preprocessing.StandardScaler()`: Chuẩn hóa dữ liệu (z-score normalization).
  - `scaler.fit_transform(X)`: Chuẩn hóa dữ liệu.
- `preprocessing.MinMaxScaler()`: Chuẩn hóa dữ liệu về khoảng [0, 1].
- `preprocessing.LabelEncoder()`: Chuyển đổi nhãn dạng chuỗi thành số.
- `preprocessing.OneHotEncoder()`: Mã hóa biến phân loại thành dạng one-hot.
- `train_test_split(X, y, test_size, random_state)`: Chia dữ liệu thành tập huấn luyện và kiểm tra.

### 5.2. Tải dataset mẫu
- `datasets.load_iris()`: Tải dataset Iris.
- `datasets.load_digits()`: Tải dataset Digits.
- `datasets.load_breast_cancer()`: Tải dataset Breast Cancer.

### 5.3. Các mô hình học máy
- **Phân loại (Classification):**
  - `LinearRegression()`: Hồi quy tuyến tính.
  - `LogisticRegression()`: Hồi quy logistic.
  - `KNeighborsClassifier()`: K-Nearest Neighbors.
  - `DecisionTreeClassifier()`: Cây quyết định.
  - `RandomForestClassifier()`: Rừng ngẫu nhiên.
  - `SVC()`: Support Vector Classifier.
- **Hồi quy (Regression):**
  - `LinearRegression()`: Hồi quy tuyến tính.
  - `Ridge()`, `Lasso()`: Hồi quy với điều chuẩn.
  - `RandomForestRegressor()`: Rừng ngẫu nhiên cho hồi quy.
- **Gom cụm (Clustering):**
  - `KMeans(n_clusters)`: Gom cụm K-Means.
  - `DBSCAN()`: Gom cụm dựa trên mật độ.
- **Giảm chiều (Dimensionality Reduction):**
  - `PCA(n_components)`: Phân tích thành phần chính (PCA).
  - `TSNE()`: T-SNE để trực quan hóa dữ liệu.

### 5.4. Huấn luyện và dự đoán
- `model.fit(X_train, y_train)`: Huấn luyện mô hình.
- `model.predict(X_test)`: Dự đoán trên tập dữ liệu mới.
- `model.score(X_test, y_test)`: Đánh giá độ chính xác (cho phân loại) hoặc R² (cho hồi quy).

### 5.5. Đánh giá mô hình
- `metrics.accuracy_score(y_true, y_pred)`: Tính độ chính xác cho phân loại.
- `metrics.precision_score()`, `metrics.recall_score()`, `metrics.f1_score()`: Tính precision, recall, F1-score.
- `metrics.confusion_matrix(y_true, y_pred)`: Tạo ma trận nhầm lẫn.
- `metrics.mean_squared_error(y_true, y_pred)`: Tính sai số bình phương trung bình (MSE).
- `metrics.r2_score(y_true, y_pred)`: Tính hệ số R².

### 5.6. Tối ưu mô hình
- `model_selection.GridSearchCV(estimator, param_grid)`: Tìm kiếm tham số tối ưu bằng lưới.
- `model_selection.cross_val_score(model, X, y, cv)`: Thực hiện kiểm định chéo (cross-validation).

### Bài 2:
- **Yêu cầu:** Tạo một mảng 1D với 20 giá trị ngẫu nhiên từ 0 đến 100, sau đó:
  1. Tính trung bình, độ lệch chuẩn, giá trị lớn nhất, và nhỏ nhất.
  2. Chuẩn hóa mảng về khoảng [0, 1] (dùng công thức:
  $x_{\text{scaled}} = \frac{x - x_{\text{min}}}{x_{\text{max}} - x_{\text{min}}}$)

  - **Hướng dẫn:** Sử dụng `np.random.randint()`, `np.mean()`, `np.std()`, `np.min()`, `np.max()`.
  
  Chuẩn hóa về cùng một khoảng [0,1], khi $x = {x_\text{min}}$ thì $x_{\text{scaled} = 0}$, $x = {x_\text{max}}$ thì $x_{\text{scaled} = 1}$ còn với các giá trị khác của x thì $x_\text{scaled} \in [0,1]$.

  Công thức mở rộng để chuẩn hóa về khoảng [$a, b$] là:  
$$x_{\text{scaled}} = a + \frac{(x - x_{\text{min}}) \cdot (b - a)}{x_{\text{max}} - x_{\text{min}}}$$

```python
import numpy as np

np_arr = np.random.randint(0, 101, 20) ## Tạo mảng 1x20
print(np_arr)

## Tính trung bình
mean = np.mean(np_arr)
print(mean)

## Tính độ lệch chuẩn
std = np.std(np_arr)
print(std)

## Tính min và max trong mảng
min = np.min(np_arr)
max = np.max(np_arr)
print(min, max)

## Chuẩn hóa mảng về 0 1
normalization_arr = np.array([(x-min)/(max-min) for x in np_arr])
print(normalization_arr)
```

  Ghép mảng theo chiều ngang, ví dụ có 2 mảng
\begin{pmatrix}
a1 & b1 & c1 & d1
\end{pmatrix}
 và
\begin{pmatrix}
a2 & b2 & c2 & d2
\end{pmatrix}
sau khi ghép theo **chiều ngang**
\begin{pmatrix}
a1 & b1 & c1 & d1 & a2 & b2 & c2 & d2
\end{pmatrix}
  **chiều dọc** thì
\begin{pmatrix}
a1 & b1 & c1 & d1 \\
a2 & b2 & c2 & d2
\end{pmatrix}

```python
import numpy

arr_1 = np.random.randint(1, 11, 5)
arr_2 = np.random.randint(1, 11, 5)
print(arr_1, arr_2)

arr = np.hstack((arr_1, arr_2)) ## Ghép mảng theo chiều dọc
print(arr)

arr_split1, arr_split2 = np.split(arr, 2) ## Chia mảng thành 2 phần bằng nhau, lưu ý số phần tử trong mảng phải chia hết cho số phần định chia
print(arr_split1, arr_split2)
```

- **Hướng dẫn:** Sử dụng `df[df[col] > value]`, `groupby()`, và `mean()`.

```python
from sklearn import datasets

iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
## Tính trung bình cột đầu tiên
mean = np.mean(iris_df.iloc[:, 0]) ## iloc[] truy cập giá trị theo index, [dòng, cột]
print(mean)
## Lọc các dòng có giá trị cột đầu tiên lớn hơn trung bình
iris_df[iris_df.iloc[:, 0] > mean]
```

```python
iris_df.groupby(iris.target).mean() ## Nhóm dữ liệu
```
